```python
import pandas as pd 
import matplotlib.pylab as plt 

x = np.arange(1, 10, 2)
y = (x / 10) + 5; 

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 10, 100) 
y = np.sin(x)

# Creating a line plot
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y = sin(x)')
plt.title('Simple Line Plot')
plt.grid()
plt.show()

plt.plot(x, y, label = 'Sine wave: label', color = 'red', marker = 's')
plt.legend();
```

```python
data = {
    'Student': ['Alice', 'Bob', 'David', 'Charlie', 'Eva'],
    'Height': [150, 160, 170, 180,190],
    'Weight': [50, 60, 65, 80, 90]
}

df = pd.DataFrame(data)
print("Data Frame : \n", df)

plt.figure(figsize = (8, 6))

plt.xlabel('Height of Students(cm)', fontsize = '12')
plt.ylabel('Weighto of Students(kg)', fontsize = '12')

plt.title('Height vs Weight of Students',fontsize = '16')
plt.scatter(df['Height'], df['Weight'],label = "Student Data", color = 'red', marker = 'o')
plt.grid()
plt.legend()
plt.show()
plt.savefig('image.png')
```

```python
import  seaborn as sns
import matplotlib.pyplot as plt

#select any numeric columns
num_df = df.select_dtypes(include=['number'])

#computer correlation matrix
corr_matrix = num_df.corr()

#plot heatmap
plt.figure(figsize= (10, 6))

sns.heatmap(corr_matrix, annot=True, cmap = "coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
