```python
data={
    'Name':['Alice','Bob','Chrlie','David',None],
    'Age':[25,np.nan,30,35,40],
    'Score':[85,90,np.nan,88,92], 
    'City':['New York','Los Angel',None, 'Chicago','Houston']
}
df=pd.DataFrame(data)
#data=pd.DataFrame(data,columns=['Name','Age','Score','City'])
display(df)
```

```python
print("Misisng val count")
display(df.isnull().sum())
```

```python
df_drop_rows=df.dropna()
print('after dropping row with missing value')
display(df_drop_rows)
```

```python
df_drop_cols=df.dropna(axis=1)
print('after dropping cols with missing value')
print(df_drop_cols,"\n")
```

```python
df_fill_mean=df.copy()
df_fill_mean['Age']= df_fill_mean['Age'].fillna(df['Age'].mean())
df_fill_mean['Score']= df_fill_mean['Score'].fillna(df['Score'].mean())
print('after fillin numeric cols with mena')
display(df_fill_mean)
```

```python
#drop column with mor ethan 20% null

nan_percentages=df.isnull().sum()/len(df)
cols_to_drop=nan_percentages[nan_percentages>.60].index
df_clean=df.drop(columns=cols_to_drop)
display(df_clean)
```

```python
df_fill_value=df.fillna({'Name':'Unknown', 'City':'Unknown','Age':0 , 'Score':0})
print("after filling With specific values")
display(df_fill_value)
```

```python
#FORWARD FILL 

df_ffill=df.ffill()
print("after forward fill")
display(df_ffill)
```

```python
df_bfill=df.bfill()
print("after backward fill")
display(df_bfill)
```

```python
df_fill_mode= df.copy()
df_fill_mode['City']= df_fill_mode['City'].fillna(df['City'].mode()[0])
print('After filling categprical columns with mode:')
display(df_fill_mode)
```
